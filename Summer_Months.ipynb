```python
import pandas as pd
import os
from glob import glob
```

```python
data_files = sorted(glob('Summer Months/citibike-*.csv'))
data_files
```

```python
final_combined_summer_2019 = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in data_files), ignore_index = True)
final_combined_summer_2019
```

```python
final_combined_summer_2019.to_csv( "final_combined_summer_2019.csv", index=False, encoding='utf-8-sig')
```
